```python
for i in range(232261):
    id, st,distr,dt,rain  = rain2023_2.iloc[i]      # here iloc and loc same
    if ds2[((ds2['State']==st) & (ds2['District']==distr) & (ds2['Date']==dt))].empty:
        fld = 0
    else:
        fld = 1
    rainfall = []
    for j in range(4):
        rainfall.append(rain2023.loc[id-j]['Avg_rainfall'])      # here iloc and loc same
    rain = np.nansum(rainfall)
    flood2023DS.loc[i] = [st,distr,dt,rain,fld]
```
